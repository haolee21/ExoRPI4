```python
import pdb
import numpy as np
atm = 101325
def LibCreate(dataList):
    data = pd.concat(dataList,ignore_index=True)
    y = np.array([data['dCyln_pre'],data['dTank_pre']]).T
    
    Theta=np.array([np.ones([len(data['tank_pre']),]),
                    data['cyln_pre']/atm,
#                     data['tank_pre']/atm,
                    data['cyln_pre']*data['tank_pre']/atm**2,
                    data['cyln_pre']/data['tank_pre'],
                    data['tank_pre']/data['cyln_pre'],
                    data['duty']/50,
                    1-data['cyln_pre']/data['tank_pre'],
                    (1-data['cyln_pre']/data['tank_pre'])**2,
                    1-data['tank_pre']/data['cyln_pre'],
                    (1-data['tank_pre']/data['cyln_pre'])**2,
                    data['cyln_pre']*data['cyln_pre']/data['tank_pre']/atm,
                    data['tank_pre']*data['cyln_pre']/data['tank_pre']/atm,
                   ]).T
    return Theta,y
```

```python
from sklearn.model_selection import train_test_split
Theta,y = LibCreate([data10_hc,data20_hc,data30_hc,data40_hc,data50_hc,data60_hc,data70_hc,data80_hc,data90_hc])
X_train, X_test, y_train, y_test = train_test_split(Theta, y, test_size=0.8, random_state=10)
```

## Solve Lasso Regression

We first train for the HC model

```python

from sklearn.linear_model import Lasso
model_hc = Lasso(alpha=0.1)
print('Number of samples: ',len(X_train))
model_hc.fit(X_train, y_train)
print('R squared training set', round(model_hc.score(X_train, y_train)*100, 2))
print('R squared testing set',round(model_hc.score(X_test,y_test)*100,2))
print(model_hc.sparse_coef_)
```

```python
Theta,y = LibCreate([data10_ht,data20_ht,data30_ht,data40_ht,data50_ht,data60_ht,data70_ht,data80_ht,data90_ht])
X_train, X_test, y_train, y_test = train_test_split(Theta, y, test_size=0.8, random_state=10)
model_ht = Lasso(alpha=0.1)
model_ht.fit(X_train, y_train)
print('Number of samples: ',len(X_train))
print('R squared training set', round(model_ht.score(X_train, y_train)*100, 2))
print('R squared testing set',round(model_ht.score(X_test,y_test)*100,2))
print(model_ht.sparse_coef_)
```
