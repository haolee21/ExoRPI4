```python
import pdb
import numpy as np
def AddTerm(termDF):
#     return termDF/np.std(termDF)
    return termDF
def LibCreate(dataList):
    data = pd.concat(dataList,ignore_index=True)
    y = np.array([data['dCyln_pre'],data['dTank_pre']]).T
    
    Theta=np.array([np.ones([len(data['tank_pre']),]),
                    AddTerm(data['cyln_pre']),
                    AddTerm(data['tank_pre']),
                    AddTerm(data['cyln_pre']*data['tank_pre']),
                    AddTerm(data['cyln_pre']/data['tank_pre']),
                    AddTerm((data['cyln_pre']/data['tank_pre'])**2),
                    data['duty'],
                   ]).T
    return Theta,y
```

```python
from sklearn.model_selection import train_test_split
Theta,y = LibCreate([data10_hc,data20_hc,data30_hc,data40_hc,data50_hc,data60_hc,data70_hc,data80_hc,data90_hc])
# Theta,y = LibCreate([data20_hc,data30_hc,data50_hc,data60_hc])
# Theta_test,y_test = LibCreate([data40_hc,data70_hc,data80_hc,data90_hc])
X_train, X_test, y_train, y_test = train_test_split(Theta, y, test_size=0.6, random_state=10)
```

## Solve Lasso Regression

```python

from sklearn.linear_model import Lasso
reg = Lasso(alpha=0.8)
reg.fit(X_train, y_train)
print('R squared training set', round(reg.score(X_train, y_train)*100, 2))
print('R squared testing set',round(reg.score(X_test,y_test)*100,2))
print(reg.sparse_coef_)
```
