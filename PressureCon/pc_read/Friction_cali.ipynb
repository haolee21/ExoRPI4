## Force sensor calibration

The force sensor needs to be calibrated everytime we use it, probably due to temperature varying in temp. 

A dry run with no force input will get us the bias voltage.

```python
data= pd.read_csv('141535_Apr122022.csv')
max_pos_v_0412 = np.mean(data['pos'])
force_bias_0412 = np.mean(data['force'])

fig,ax = plt.subplots(2,1,figsize=(15,10))
ax[0].plot(data['time'],data['force'])
ax[0].set_ylabel('Force')
ax[1].plot(data['time'],data['pos'])
ax[1].set_ylabel('Position')
```
